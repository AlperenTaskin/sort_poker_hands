```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import operator
def sort_numbers(df):
    
    df = pd.DataFrame(df)
    new_df = df.copy()
    C_S_map = {'C1':'S1','C2':'S2','C3':'S3','C4':'S4','C5':'S5'}
    
    for index,row in new_df.iterrows():
        
        liste = list()
        C_liste = list()
        S_liste = list()

        for item in row:
            liste.append(item)

        # S ve C lerin bulundugu listeler
        S_liste = liste[5:]    
        C_liste = liste[:5]

        # Yer degistirdiklerinde C'lerin karsisindaki S'leri bulabilmek icin dictionary'e yaziyoruz
        Cler = {'C1':C_liste[0],'C2':C_liste[1],'C3':C_liste[2],'C4':C_liste[3],'C5':C_liste[4]}
        C_siralanmis = sorted(Cler.items(), key=operator.itemgetter(1))

        Sler = {'S1':S_liste[0],'S2':S_liste[1],'S3':S_liste[2],'S4':S_liste[3],'S5':S_liste[4]}

        # C'ler siralandiktan sonra S'leri de map listesi yardimiyla siraliyoruz
        S_siralanmis = list()
        for i in range(len(Sler)):
            S_siralanmis.append(Sler[C_S_map[C_siralanmis[i][0]]])
        
        
        for i in range(len(C_siralanmis)):
            new_df.set_value(index, X_train.columns[i], C_siralanmis[i][1])
            new_df.set_value(index, X_train.columns[i+5], S_siralanmis[i])
        
    return new_df
```

### Tum Veri Setini Siralama 

```python
new_X_train = X_train.copy()
```

```python
new_X_train = sort_numbers(new_X_train)
```

```python
new_X_train.to_csv('sorted_poker.csv', index=False)
```
